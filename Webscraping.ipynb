```python
 # Importamos el módulo para descargar las páginas web
import urllib.request
```

```python
# Definimos la cabecera para que no rechace la solicitud
header= {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) ' 
      'AppleWebKit/537.11 (KHTML, like Gecko) '
      'Chrome/23.0.1271.64 Safari/537.11',
      'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
      'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
      'Accept-Encoding': 'none',
      'Accept-Language': 'en-US,en;q=0.8',
      'Connection': 'keep-alive'}

# Definimos la página y realizamos la consulta
url = 'https://autotest.com.ar/pruebas/test-drive-renault-duster-iconic/'
req = urllib.request.Request(url=url, headers=header) 
data = urllib.request.urlopen(req).read().decode()
```

```python
data
```

```python
# Extraer los datos con pandas
import pandas as pd
dfs = pd.read_html(data)
confort = dfs[0]
confort
```

```python
# Fusionamos los datos para armar una ficha técnica
ficha_tecnica = pd.DataFrame()
for i in range(6,10):
  ficha_tecnica = ficha_tecnica.append(dfs[i])
ficha_tecnica
```

```python
#Extracción de datos con BeatifulSoup
from bs4 import BeautifulSoup
soup =  BeautifulSoup(data, 'html.parser')
for i in range(2,6):
  imagenes = soup.find_all('img')[i]
  print(imagenes)
```

```python
videos = soup.find_all('iframe')
print(videos)
```
